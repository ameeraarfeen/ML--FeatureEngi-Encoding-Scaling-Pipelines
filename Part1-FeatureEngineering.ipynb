### task 1- creating new features

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cleaned_titanic.csv")
```

```python
df.head()
```

```python
#feature for family size 

df['family_size']=df["parch"]+df["sibsp"]+1
```

```python
#is_child feature creation
df['is_child']=np.where(df['age']<12 , 1, 0)
```

```python
df[['age', 'sibsp', 'parch','family_size','is_child']].head()
```

### task 2- handeling date and text features

```python
df.dtypes
```

```python
df['title']=df['name'].str.extract(r' ([A-Za-z]+)\.')
```

```python
df['title'].value_counts()
```

```python
df['title'] = df['title'].apply(lambda x: x if x in ['Mr','Miss','Mrs','Master'] else 'Other')
```

```python
df['title'].value_counts()
```

```python
df.to_csv('featured_titanic.csv', index=False)
```
