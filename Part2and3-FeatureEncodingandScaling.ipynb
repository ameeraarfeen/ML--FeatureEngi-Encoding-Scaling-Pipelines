### Task 4 - Column Transformer 

```python
numerical_cols = ['age', 'fare', 'family_size', 'pclass', 'is_child', 
                  'sex_encoded', 'embarked_C', 'embarked_Q', 'embarked_S',
                  'title_Master', 'title_Miss', 'title_Mr', 'title_Mrs', 'title_Other']
categorical_cols = []  # empty because already encoded
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[
    ('num', 'passthrough', numerical_cols),
    ('cat', OneHotEncoder(), categorical_cols)
])
```

```python
X = df.drop(columns=['survived', 'passengerid'])
y = df['survived']

ct_transformed = ct.fit_transform(X)
print(ct_transformed.shape)
```

## Part3 Feature Scaling

### Task 5- Standardisation (Standard Scaler)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() #creating an object -StandardScaler is a class, and we need an instance of it before using its method
```

```python
#Since you're both: learning the mean/std from the data (fit)
#and transforming the data (transform)
scaled_data=scaler.fit_transform(df[numerical_cols])
```

```python
scaled_df = pd.DataFrame(
    scaled_data,
    columns=numerical_cols
)
```

```python
scaled_df.mean()
```

```python
scaled_df.std()
```

### Task 6- Normalisation (MinMaxScaler)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minMaxScaler = MinMaxScaler()
```

```python
minMaxScaled_data=minMaxScaler.fit_transform(df[numerical_cols])
```

```python
minMaxScaled_df = pd.DataFrame(
    minMaxScaled_data,
    columns=numerical_cols
)
```

```python
minMaxScaled_df.min()
```
